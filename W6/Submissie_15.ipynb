4. Schrijf een test die checked dat die kansen optellen tot 1.
5. **Advanced** In de een na laatste paragraaf van 3.3 op blz 76 wordt een andere manier om die kansen te berekenen gemeld, die dezelfde uitkomst geeft. Programmeer die ook, en vergelijk de uitkomsten.
    

```python
Fig35=nx.Graph()
nodes = ['mary', 'nancy', 'tom', 'john', 'pam', 'bob', 'tara']
blue_node = ['tom']
Fig35.add_nodes_from(nodes)

Fig35.add_edges_from([('mary', 'nancy'), ('nancy', 'tom'), ('nancy', 'bob'), ('bob', 'tara'), ('tom', 'pam'),
                     ('john', 'pam'), ('tom', 'john'),('tom', 'bob')])
for node in Fig35: 
    if not node in blue_node:
        Fig35.nodes[node]['B'] = False
        Fig35.nodes[node]['R'] = True
    else:
        Fig35.nodes[node]['B'] = True 
        Fig35.nodes[node]['R'] = False

color_map = []
for node in Fig35:
    if Fig35.nodes[node]['R'] == False:
        color_map.append('blue')
    else: 
        color_map.append('red')      
nx.draw(Fig35, node_color=color_map, with_labels=True)
plt.show()
        

def kans_op_knoop(G,knoop):
    #WRITE YOUR CODE HERE
    kans = []
    for i in G:
        if G.has_edge(knoop,i):
            kans.append((1/G.degree(i)) * (1/nx.number_of_nodes(G)))
        else:
            kans.append(0)
    return sum(kans)

 

H = nx.karate_club_graph() 
# test
sum(kans_op_knoop(H,x) for x in H)
#kans_op_knoop(H,1)
```

```python
assert isinstance(Fig35, nx.Graph)
H= nx.karate_club_graph()
assert 0 <= kans_op_knoop(H,1)<= 1
```
